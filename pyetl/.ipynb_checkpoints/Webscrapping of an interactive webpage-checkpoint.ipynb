```python
import requests
# json for interactive webpage
import json
from bs4 import BeautifulSoup
url ="https://buzzorange.com/techorange/wp/wp-admin/admin-ajax.php"
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
# this is obtianed from payload section and used to give to the data variable of the response
datastr ='''action: ceris_posts_listing_grid_b
args[cat]: 396
args[post__not_in]: 
args[post_type][]: post
args[posts_per_page]: 100
args[post_status]: publish
args[offset]: 0
args[orderby]: date
postOffset: 200 # this one is the variable that the API uses to get new view
type: loadmore
moduleInfo[post_source]: all
moduleInfo[post_icon]: enable
moduleInfo[post_icon_animation]: enable
moduleInfo[iconPosition]: top-right
moduleInfo[bookmark]: off
securityCheck: 8bfabd4c8e'''
data ={ row.split(': ')[0]:row.split(': ')[1] for row in datastr.split("\n")}
response = requests.post(url,headers=headers,data=data)
json_data = json.loads(response.text)
soup = BeautifulSoup(json_data,'html.parser')
soup.select('div')
```
