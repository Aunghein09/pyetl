```python
import jieba
# glob to accesss the files inside a dir
import glob
```

```python
glob.glob("./res_gossiping/*.txt") # return a list
```

```python
sb = ''
for filepath in glob.glob("./res_gossiping/*.txt"):
#     with open(filepath,'r',encoding='utf-8') as fh:
#         sb += fh.read().split('---split---')[0]
    sb += open(filepath,'r', encoding='utf-8')\
            .read()\
            .split("---split---")[0]
sb
```

```python
# this will only return a generator
word_generated = jieba.cut(sb)
# we will save the words in a list
word_ls =[]
for word in word_generated:
    print(word)
    word_ls.append(word)
```

```python
word_ls
```

```python
word_counts = {}
for word in word_ls:
    word_counts[word] = word_counts.get(word,0) + 1
```

```python
# see the cell below, items give back tuple and [1] is to get the vlaue
test = sorted(word_counts.items(), key= lambda kv: kv[1], reverse=True)
test
```

```python
# using k,v
wc_list = [(k,v) for k,v in word_counts.items()]
wc_list.sort(key=lambda item: item[1], reverse=True)
wc_list
```

```python
# laoshi's code
jieba.load_userdict('mydict.txt')
sb = ''
for filepath in glob.glob('./res_gossiping/*.txt'):
    sb += open(filepath, 'r', encoding='utf-8') \
            .read() \
            .split("---split---")[0]
    
wc_generator = jieba.cut(sb)
wc_list = list()
for w in wc_generator:
#     print(w)
    wc_list.append(w)
wc_map = dict()
for w in wc_list:
    if w in wc_map:
        wc_map[w] += 1
    else:
        wc_map[w] = 1
wc_sort_list = [(k, v) for k, v in wc_map.items()]
wc_sort_list.sort(key=lambda x: x[1], reverse=True)
wc_sort_list
```

```python
jieba.load_userdict('mydict.txt')
stopwords = open('stopword.txt','r',encoding='utf-8').readlines()
stopwords = [i.replace("\n","") for i in stopwords]
stopwords.append('\n') 
sb = ''
for filepath in glob.glob('./res_gossiping/*.txt'):
    sb += open(filepath, 'r', encoding='utf-8') \
            .read() \
            .split("---split---")[0]
    
wc_generator = jieba.cut(sb)
wc_list = list()
for w in wc_generator:
#     print(w)
    wc_list.append(w)
wc_map = dict()
for w in wc_list:
    if w in stopwords or len(w)<2:
        continue
    if w in wc_map:
        wc_map[w] += 1
    else:
        wc_map[w] = 1
wc_sort_list = [(k, v) for k, v in wc_map.items()]
wc_sort_list.sort(key=lambda x: x[1], reverse=True)
wc_sort_list
```
